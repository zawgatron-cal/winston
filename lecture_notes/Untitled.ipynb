```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)
```

```python
#make a figure
plt.figure(figsize=(8,8))

#do the plotting
plt.scatter(x, y)
plt.plot(x, y**3)

#labels
plt.xlabel("x")
plt.ylabel("y")

# show it
plt.show()
```
